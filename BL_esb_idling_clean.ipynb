```python
df = pd.read_csv('esb_cleaned.csv')
```

# Data Cleaning

```python
df_clean = df.drop(['id','Vehicle ID', 'Fleet ID', 'Model Year', 'Initial Service Year', 'Initial SOC',	'Final SOC', 'Min SOC',	'Max SOC','% Electrified', 'Data Collection End', 'Data Collection Start', 'Model Name', 'Manufacturer', 'Body Style', 'Sector', 'Vocation', 'Vehicle Platform', 'Fleet Type', 'Data Collection Method', 'vehicle_name', 'Curb Weight', 'Capital Cost', 'Nominal Range', 'Nominal Efficiency', 'Peak Power', 'Peak Power', 'Towing Capacity', 'Max Charge Rate', 'State', 'Census Region', 'Battery Chemistry', 'Peak Torque', 'Weight Class', 'Charge Connector', 'Heater Type', 'Rated Energy'], axis=1)
df_clean.head()
```

```python
df_clean['Region'].unique()
```

```python
df_clean['Total Fleet Size'].unique()
```

```python
#creating region dummy column
df_clean.loc[df_clean['Region'] == 'NC', 'Region_NC'] = 1
df_clean['Region_NC'] = df_clean['Region_NC'].fillna(0)
df_clean.loc[df_clean['Region'] == 'SC', 'Region_SC'] = 1
df_clean['Region_SC'] = df_clean['Region_SC'].fillna(0)
df_clean.loc[df_clean['Region'] == 'RM', 'Region_RM'] = 1
df_clean['Region_RM'] = df_clean['Region_RM'].fillna(0)
#add a over65 column where 
df_clean.loc[df_clean['Fuel Fired Heater?'] == False, 'Elec_heat'] = 0
df_clean.loc[df_clean['Fuel Fired Heater?'] == True, 'Elec_heat'] = 1

#create a new column for energy per mile
df_clean.loc[df_clean['Average Ambient Temperature'] > 65, 'Over65'] = 1
df_clean.loc[df_clean['Average Ambient Temperature'] <= 65, 'Over65'] = 0
df_clean.head()
```

```python
df_numeric = df_clean.drop(['Fleet', 'Date', 'Fleet_full', 'Region', 'Total Fleet Size','Fuel Fired Heater?', 'Average Ambient Temperature'], axis=1)
df_numeric.rename(columns={'Energy Consumption Rate': 'energy_consumption_rate', 'Total Distance': 'distance', 'Total Run Time': 'run_time', 'Idling Time': 'idling_time', 'Driving Time':'driving_time', 'SOC Used':'soc_used', 'Total Energy Consumption':'energy_consumed', 'Average Speed': 'average_speed', 'Initial Odometer':'initial_odometer', 'Final Odometer':'final_odometer', 'Percent Idling Time':'percent_idling_time'}, inplace=True)
df_numeric.head(1)
```

```python
corr = df_numeric.corr()
corr.style.background_gradient(cmap='coolwarm')
```

# Base Regression

```python
model = smf.ols(formula='energy_consumed ~  distance + idling_time + soc_used + average_speed + Region_NC + Region_SC + Region_RM  + Elec_heat', data=df_numeric).fit()
# # model summary
print(model.summary())
```
